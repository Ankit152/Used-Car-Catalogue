**The dataset contains missing vales**

```python
data['engine_capacity']=data['engine_capacity'].fillna(data['engine_capacity'].mean())
data.isna().sum()
```

```python
col_dummy=[]
col_std=[]
y=data['engine_type']
data=data.drop('engine_type',axis=1)
for col in list(data.columns):
    n=len(pd.unique(data[col]))
    print("Number of unique elements in the column ",col," is: ",n)
    if(n<13):
        col_dummy.append(col)
    else:
        col_std.append(col)
        
print("<<<==============>>>")
print("Number of columns for creating dummy variables: ",len(col_dummy))
print("Number of columns for standardisation is: ",len(col_std))
```

```python
# creating the dummy variables
data=pd.get_dummies(data[col_dummy])
data.head()
```
